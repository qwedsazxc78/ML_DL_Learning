```python
def fast_plot_line(figure, axis, index, time , x, y, z, length ,title, ylimit):
    axis[index].plot()
    axis[index].plot(time, x, label='magX')
    axis[index].plot(time, y, label='magY')
    axis[index].plot(time, z, label='magZ')
    axis[index].plot(time, length, label='magLen')
    axis[index].set_xlabel('frameCnt')
    axis[index].set_ylabel('Mag Value')
    axis[index].set_title(title)
    axis[index].legend()
    axis[index].grid()
    if ylimit:
        axis[index].set_ylim([-100, 100])
    return None

def df_process(df):
    df['mag_len'] = (df['magX'].pow(2) + df['magY'].pow(2) + df['magZ'].pow(2)).pow(1./2)
    df['magX_WO_offset'] = df.magX - df.magX[0]
    df['magY_WO_offset'] = df.magY - df.magY[0]
    df['magZ_WO_offset'] = df.magZ - df.magZ[0]
    df['mag_len_WO_offset'] = (df['magX_WO_offset'].pow(2) + df['magY_WO_offset'].pow(2) + df['magZ_WO_offset'].pow(2)).pow(1./2)
    return df

def df_Pos_process(df, spf1, spf1str):
    df[str(spf1str) +'_X'] =  spf1['magX'].mean()
    df[str(spf1str) +'_Y'] =  spf1['magY'].mean()
    df[str(spf1str) +'_Z'] =  spf1['magZ'].mean()
    df[str(spf1str) +'_Y'] =  spf1['magT'].mean()
    return df
```
